```python
def calculateReconstructionError(X, X_hat):
    return np.mean(np.sum(np.square(X - X_hat), axis=1))
```

## Function to plot confusion matrix

```python
def plotConfusionMatrix(y_true, y_pred, title):
    sns.set_style("white")
    confusionmat = ConfusionMatrixDisplay.from_predictions(y_true, 
                                        y_pred,
                                       cmap=plt.cm.Reds)
    confusionmat.ax_.set_xlabel("Predicted Label", fontsize=12, weight="bold")
    confusionmat.ax_.set_ylabel("True Label", fontsize=12, weight="bold")
    confusionmat.ax_.set_title(title, fontsize=15, weight="bold")
    plt.show()
```

# Decided to use PCA

```python
pca = PCA()
pca.fit(X_train)
```

```python
count = 0
for var in np.cumsum(pca.explained_variance_ratio_):
    count+=1
    if var >= 0.95:
        print(count)
        break
```

```python
transformed_X_test = pca.transform(X_test)[:, :count]
transformed_X_train = pca.transform(X_train)[:, :count]
```

## Function to plot ROC curve

```python
def plot_ROC(Y, predicted_proba, rocauc_score):
    """
    Y: Y_train
    predicted_probe: the predicted probabilities returned by sklearn's predict_proba
    rocauc_score: the macro average ovo rocauc_score returned by sklearn's roc_auc_score
    return:
    the figure
    """
    fpr_grid = np.linspace(0.0, 1.0, 1000)
    label_binarizer = LabelBinarizer().fit(Y)
    pair_list = list(combinations(np.unique(Y), 2))
    pair_scores = []
    mean_tpr = dict()
    colors = [sns.color_palette("Dark2")[i] for i in range(len(pair_list))]
    for ix, (label_a, label_b) in enumerate(pair_list):
        a_mask = Y_test == label_a
        b_mask = Y_test == label_b
        ab_mask = np.logical_or(a_mask, b_mask)

        a_true = a_mask[ab_mask]
        b_true = b_mask[ab_mask]

        idx_a = np.flatnonzero(label_binarizer.classes_ == label_a)[0]
        idx_b = np.flatnonzero(label_binarizer.classes_ == label_b)[0]

        fpr_a, tpr_a, _ = roc_curve(a_true, predicted_proba[ab_mask, idx_a])
        fpr_b, tpr_b, _ = roc_curve(b_true, predicted_proba[ab_mask, idx_b])

        mean_tpr[ix] = np.zeros_like(fpr_grid)
        mean_tpr[ix] += np.interp(fpr_grid, fpr_a, tpr_a)
        mean_tpr[ix] += np.interp(fpr_grid, fpr_b, tpr_b)
        mean_tpr[ix] /= 2
        mean_score = auc(fpr_grid, mean_tpr[ix])
        pair_scores.append(mean_score)
    
    ovo_tpr = np.zeros_like(fpr_grid)
    fig, ax = plt.subplots(figsize=(6, 6))
    for ix, (label_a, label_b) in enumerate(pair_list):
        ovo_tpr += mean_tpr[ix]
        plt.plot(
            fpr_grid,
            mean_tpr[ix],
            label=f"Mean {label_a} vs {label_b} (AUC = {pair_scores[ix]:.2f})",
            color=colors[ix]
        )

    ovo_tpr /= sum(1 for pair in enumerate(pair_list))

    plt.plot(
        fpr_grid,
        ovo_tpr,
        label=f"One-vs-One macro-average (AUC = {rocauc_score:.2f})",
        linestyle=":",
        linewidth=4,
        color="red"
    )
    plt.plot([0, 1], [0, 1], "k--", label="Chance level (AUC = 0.5)")
    plt.axis("square")
    plt.xlabel("False Positive Rate", fontsize=12, weight="bold")
    plt.ylabel("True Positive Rate", fontsize=12, weight="bold")
    plt.title("One-vs-One Multiclass ROC", fontsize=12, weight="bold")
    plt.legend()
    plt.show()
    return fig
```

```python
rf_classifier = RandomForestClassifier()
param_grid_rf = {"n_estimators": np.arange(50, 1050, 150),
                "max_depth": [10, 30, 50, 70, 90, 110, 130, None],
                "min_samples_split": np.arange(2, 10, 2),
                 "min_samples_leaf": np.arange(1, 10, 2),
                 "max_features": ["sqrt", "log2"]}
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=random_seed)
rf_grid_search = RandomizedSearchCV(rf_classifier, param_grid_rf, n_iter=60, scoring='f1_macro', n_jobs=-1, cv=cv, refit=True, verbose=True)
```
